```python
import numpy as np
import pandas as pd
import re
import string
import pickle
```

```python
txt='great product. i like it'
```

```python
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, ' ')
    return text
```

```python
with open('../static/model/corpora/stopwords/english', 'r') as file:
    sw = file.read().splitlines()
```

```python
def preprocessing(text):
    data = pd.DataFrame([text], columns=['tweet'])

    # Lowercase the text
    data["tweet"] = data["tweet"].apply(lambda x: " ".join(word.lower() for word in x.split()))

    # Remove URLs
    data["tweet"] = data["tweet"].apply(lambda x: re.sub(r'http\S+', '', x))

    # Remove punctuations
    data["tweet"] = data["tweet"].apply(remove_punctuations)

    # Remove digits
    data["tweet"] = data["tweet"].str.replace(r'\d+', '', regex=True)

    # Remove stopwords
    data["tweet"] = data["tweet"].apply(lambda x: " ".join(word for word in x.split() if word not in sw))

    # Apply stemming
    data["tweet"] = data["tweet"].apply(lambda x: " ".join(ps.stem(word) for word in x.split()))

    return data["tweet"]
```

```python
preprosesed_txt = preprocessing(txt)
```

```python
vocab = pd.read_csv('../static/model/vocabulary.txt', header=None)
tokens = vocab[0].tolist()
```

```python
tokens
```

```python
def vectorizer(ds, vocabulary):
    vectorized_lst = []
    
    for sentence in ds:
        sentence_lst = np.zeros(len(vocabulary))
        
        for i in range(len(vocabulary)):
            if vocabulary[i] in sentence.split():
                sentence_lst[i] = 1
                
        vectorized_lst.append(sentence_lst)
        
    vectorized_lst_new = np.asarray(vectorized_lst, dtype=np.float32)
    
    return vectorized_lst_new
```
